```python
inputs = []
outputs = []
for word in data["content"][0:200]:
    for i in range(len(token.texts_to_sequences([word])[0]) -1):
        inputs.append(token.texts_to_sequences([word])[0][:i+1])
        outputs.append(token.texts_to_sequences([word])[0][i+1])
```

```python
from keras.utils import pad_sequences
```

```python
fv = pad_sequences(inputs)
```

```python
from keras.utils import to_categorical
```

```python
cv = to_categorical(outputs,num_classes = 951)
```

```python
cv
```

```python
cv.shape
```

```python
from keras.models import Sequential
```

```python
from keras.layers import Dense,LSTM,Embedding,Input
```

```python
model = Sequential()
```

```python
model.add(Embedding(951, 10, input_shape = (114,)))
model.add(LSTM(200, activation ="tanh", return_sequences=False ))
model.add(Dense(951,activation = "softmax"))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss= "categorical_crossentropy", metrics = ["accuracy"])
```

```python
model.fit(fv, cv, epochs= 100, batch_size = 50, validation_split = 0.2)
```

```python
import numpy as np
from tensorflow.keras.preprocessing.sequence import pad_sequences

text = "Please"
for i in range(5):
    # Tokenize the input text
    sequences = token.texts_to_sequences([text])

    # Pad the sequences to ensure consistent input size for the model
    padded_sequences = pad_sequences(sequences, maxlen=114)

    # Predict the next word
    predicted_probs = model.predict(padded_sequences)
    v = np.argmax(predicted_probs)

    # Get the word corresponding to the predicted index
    predicted_word = token.index_word.get(v, '')  # Use .get to handle out-of-vocabulary indices

    # Append the predicted word to the text
    text = text + " " + predicted_word
    print(text)
```
